# Matplotlib em attacks_pattern

```python
import pandas as pd

df = pd.read_csv('attacks_pattern.csv')
df.head()
```

```python
df['cum_counter_attack'] = df['is_counter_attack'].cumsum()
df.tail()
```

## Utilização Matplot lib

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot(df['timestamp'], df['cum_counter_attack'], label='Usuário')
ax.plot(df['timestamp'], df['damage'], label='Naïve Bayes')

ax.set_title('Comparação entre Inputs de Usuário vs Gerado pelo modelo Naïve Bayes')
ax.set_ylabel('Contra-ataques Acumulados')
ax.legend(title='Origem')
```
